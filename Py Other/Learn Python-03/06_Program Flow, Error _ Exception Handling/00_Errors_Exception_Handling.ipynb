The basic structure of TRY-EXCEPT statement looks like:

try: 
    {code blocak to be normally executed
    
except:
    {code block to be exceptionally executed

Kod burada çalışırsa sorun yok devam eder ama

```python
while True :
    no_one = int(input("the first number please :"))
    no_two = int(input("the second number please :"))
    
    try :
        division = no_one / no_two    # Bady de hata yoksa Try nloğunu çalıştır.
        print("the result of the division is : ", division)
        break
        
    except : # Bodyde hata varsa ben buradayım. gel beni çalıştır. demiş olduk.
        print("something went wrong...Try again.")
        # break
```

```python
while True :
    no_one = int(input("the first number please :"))
    no_two = int(input("the second number please :"))
    
    try :
        division = no_one / no_two

    except ZeroDivisionError: # ZeroD.. hatası varsa beni çalıştır.
        print("You can't divide by zero!...Try again.")

    else :  # Try doğruysa elseyi çalıştır.
        print("the result of the division is : ", division)

    finally :   # beni her halukarda çalıştır.
        print("thanks for using our mini divison calculator. come again!")
        break
```
