```python
from itertools import count
for i in count(3, 2):  # işlev sayımı, bir değerden sonsuza kadar saymaya yarar.
  print(i)
  if i >=11:
    break
```

```python
from itertools import takewhile

nums = [2, 4, 6, 8, 10, 7, 9, 8]

a = takewhile(lambda x: x%2==0, nums) # sadece ardışıkları
print(list(a))
```

```python
from itertools import product, permutations

letters = ("A", "B")
print(list(product(letters, range(2))))
print(list(permutations(letters)))
```
