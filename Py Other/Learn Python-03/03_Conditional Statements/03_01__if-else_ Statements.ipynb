Comparison Operators
Koşul Boolen değer vermesi için işin içine aşağıdaki karşılaştırma operatörleri girmeli

== Returns True if two values are equal or False if different 2 == 2 (True), 2 == 3 (False)

!= Returns True if two values are not equal or False if equal 2 != 2 (False), 2 != 3 (True)

büyük -- Returns True if the value on the left is greater than the value on the right otherwise returns False 3 > 2 (True), 2 > 3 (False)

küçük -- Returns True if the value on the left is less than the value on the right otherwise returns False 2 < 3 (True), 3 < 2 (False)

büyük =  ---3 >= 2 (True), 3 >= 3 (True), 2 >= 3 (False)

<=       ---3 <= 2 (False), 3 <= 3 (True), 2 <= 3 (True)

```python
x = 6
y = 9
print ("is x equal to y?                 :" , x == y)
print ("is x not equal to y?             :" , x != y)
print ("is x less than y?                :" , x < y)
print ("is x greater than y?             :" , x > y)
print ("is x less than or equal to y?    :" , x <= y)
print ("is x greater than or equal to y? :" , x >= y)
```
