```python
a = int(input('pozitif sayı ile giriş yapın'))
b = int(input('pozitif sayı ile giriş yapın'))

('Bu sayılar eşit' if a == b else "Negatif giriş yaptınız" ) if (a < 0 or b < 0) else (a - b if a > b  else b - a)
```

Sonrasında büyük sayıyı küçük sayıdan çıkaran tek satırlık ternary ile sorunun çözümünü tamamlayabiliriz.


```python
while True :
   a = int(input('pozitif sayı ile giriş yapın: '))
   b = int(input('pozitif ve ilk sayıdan farklı bir giriş yapın: '))
   if not (a == b or a < 0 or b < 0) :
     break
a - b if a > b else b - a
```
