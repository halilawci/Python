```python
def calculator(x, y, opr) :
    if opr == "+" :
        print(x + y)
    elif opr == "-" :
        print(x - y)
    elif opr == "*" :
        print(x * y)
    elif opr == "/" :
        print(x / y)
    else :
        print("enter a valid operator")
calculator(22, 11, "/")        
```

Task : mutlak değer yazan bir fonksiyon ve bir doc yazın

```python
def absolute_value(num) :
    """This function returns the 
    absolute value of the given number"""
    if num > 0 :
        return num
    
    else :
        return -num
```

```python
absolute_value(-55)
```

```python
print(absolute_value.__doc__)
```
