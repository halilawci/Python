(Nesneye yönelik planlama)OOP'nin Temel İlkeleri
OOP'nin dört temel ilkesi vardır. Bunlar  kapsülleme, soyutlama, kalıtım  ve  polimorfizmdir. 

Veri kapsülleme  , nesnelerin iç verilerini dünyadan gizleme mekanizmasıdır. Nesne ve verileriyle olan tüm etkileşimler, genel yöntemleri aracılığıyla gerçekleştirilir. Kapsülleme, programcıların nesneyi tutarsızlıktan korumasını sağlar.
Veri soyutlama  , nesnelerin uygulamalarının basitleştirilmiş, soyut versiyonunu sağlaması gerektiği anlamına gelir. Dahili çalışmalarının ayrıntıları genellikle kullanıcı için gerekli değildir, bu nedenle onları temsil etmeye gerek yoktur. Soyutlama aynı zamanda nesnenin yalnızca en alakalı özelliklerinin sunulacağı anlamına gelir.
Kalıtım  , sınıflar arasındaki ebeveyn-çocuk ilişkilerini tanımlamak için bir mekanizmadır. Genellikle nesneler çok benzerdir, bu nedenle kalıtım, programcıların ortak mantığı yeniden kullanmalarına ve aynı zamanda sınıflara benzersiz kavramlar getirmelerine olanak tanır.
Polimorfizm  , kelimenin tam anlamıyla  bir isim ve birçok form anlamına gelir  ve sınıfların kalıtımı ile ilgilidir. Adından da anlaşılacağı gibi, programcıların aynı yöntemin farklı mantığını tanımlamasına izin verir. Yani ad (veya arayüz) aynı kalır, ancak gerçekleştirilen eylemler farklı olabilir. Uygulamada, aşırı yükleme veya geçersiz kılma ile yapılır.

```python
class Person:  # genellikle class isimleri büyük harfle başlar.
    # class attributes
    adress = 'no information'

    # constructor (yapıcı metod); constroctor tanımlaması yapabilmek için aşağıdaki yolu izlemeliyiz.
    def __init__(self, name, year):  # init metodunun ilk parametresi self'dir. (self yerine başka birşeyde yazılabilir.) self class'dan türetilen p1, p2 objelerini temsil eder. obje üzerine değer atamak için self kullanılacak. burada name, surname vb atandı.
      # object attributes
      self.name = name
      self.year = year

    # instance methods
    def intro(self):
      print('hello' + self.name)
    # instance methods
    def calculateAge(self):
      return 2022- self.year

# object (instance)
p1 = Person(name = 'ali', year = 1990)  # obje genellikle küçük harfle başlar. sayısal ifade ile başlamaz.
p2 = Person(name ='sena', year =1986)   # paramaetrelerin isimleri yani name, year yazmasakta olur.

p1.intro()
p2.intro()

print(f'adım : {p1.name} ve yaşım: {p1.calculateAge()}')
print(f'adım : {p2.name} ve yaşım: {p2.calculateAge()}')
```

```python
class Circle:
  # class attributes
  pi = 3.14

  def __init__(self, yaricap=1):
    self.yaricap = yaricap

  # methods
  def cevre_hesaplama(self):
    return 2*self.pi + self.yaricap

  def alan_hesaplama(self):
    return self.pi * (self.yaricap**2)

c1 = Circle()  #yaricap = 1
c2 = Circle(5)

print(f'c1: alan = {c1.alan_hesaplama()} c1: çevre = {c1.cevre_hesaplama()}')
print(f'c2: alan = {c2.alan_hesaplama()} c2: çevre = {c2.cevre_hesaplama()}')
```

İnheritance (Kalıtım): Classlar arası miras.

```python
# Person => name, lastname,  age, eat(), run(), drink() # person class'ına method  eklemiş olduk.
# student(Person), Teacher(Person) Person class ında olan tüm attribut ve methotları student ve teacher'e de ekledik.

class Person():
  def __init__(self, fname, lname):
    self.firstName = fname
    self.lastName = lname
    print('person Created')

  def who_am_i(self):
    print('I am a person')
  def eat(self):
    print('I am eating')

class Student(Person):
  def __init__(self, fname, lname, number):
    Person.__init__(self, fname, lname)
    self.studentNumber = number
    print('Student Created')

  def who_am_i(self):
    print('I am a student')

  def sayHello(self):
    print('Hello I am a student')



class Teacher(Person):
  def __init__(self, fname, lname, branch):
    super().__init__(fname, lname)
    self.branch = branch

  def who_am_i(self):
    print(f'I am a {self.branch} teacher')

p1 = Person('ali', 'yılmaz')
s1 = Student('Çınar', 'Turan', 1256)
t1 = Teacher('Serkan', 'Yılmaz', 'Matematik')

print(p1.firstName+ ' ' + p1.lastName)
print(s1.firstName+ ' ' + s1.lastName+ ' ' + str(s1.studentNumber))

p1.who_am_i()
s1.who_am_i()
p1.eat()
s1.eat()
s1.sayHello()
t1.who_am_i()


```

```python
# Question

class Question:
  def __init__(self, text, choices, answer):
    self.text = text
    self.choices = choices
    self.answer = answer

  def checkAnswer(self, answer):
    return self.answer == answer
# Quiz
class Quiz:
  def __init__(self, questions):
    self.questions = questions
    self.score = 0
    self.questionIndex = 0

  def getQuestion(self):
    return self.questions[self.questionIndex]

  def displayQuestion(self):
    question = self.getQuestion()
    print(f'Soru {self.questionIndex +1} : {question.text}')
    for q in question.choices:
      print('-' + q)

    answer = input('cevap: ')
    self.guess(answer)
    self.loadQuestion()

  def guess(self, answer):
    question = self.getQuestion()
    if question.checkAnswer(answer):
      self.score += 1
    self.questionIndex += 1
    self.displayQuestion()
  def loadQuestion(self):
    if len(self.questions) == self.questionIndex:
      self.showScore()
    else:
      # self.displayProgress()
      self.displayQuestion()

  def showScore(self):
    print('score: ', self.score)

  def displayProgress(self):
    totalQuestion = len(self.questions)
    questionNumber = self.questionIndex + 1
    if questionNumber > totalQuestion:
      print('Quiz bitti.')
    else :
      print(f"Question{questionNumber} of {totalQuestion}". center(100, '*'))


q1 = Question('en iyi dil hangisi?', ['C#', 'python', 'javascript', 'java'], 'python')
q2 = Question('en popüler dil hangisi?', ['C#', 'python', 'javascript', 'java'], 'python')
q3 = Question('en kazandıran dil hangisi?', ['C#', 'python', 'javascript', 'java'], 'python')
questions = [q1, q2, q3]

quiz = Quiz(questions)
quiz.loadQuestion()




```
