```python
class calisan:
  def __init__(self, adi, soyadi, maasi, yasi):
    self.adi = adi  # Bunlar attribute
    self.soyadi = soyadi
    self.maasi = maasi
    self.yasi = yasi
  def show_info(self):
    print(f"Ad: {self.adi} soyadı: {self.soyadi} maaşı: {self.maasi}")

calisan1 = calisan('ali', 'demir', 5000, 32)
print(calisan1.adi, calisan1.soyadi, calisan1.maasi, calisan1.yasi)
calisan2 = calisan('ahmet', 'yaprak', 3500, 62)
print(calisan2.adi, calisan2.soyadi, calisan2.maasi, calisan2.yasi)

calisan1.show_info()
calisan2.show_info()
```

```python
class calisan:
  def __init__(self, adi, soyadi, maasi, yasi):
    self.adi = adi  # Bunlar attribute
    self.soyadi = soyadi
    self.maasi = maasi
    self.yasi = yasi
  def show_info(self):  # claslara ait fonksiyonlara methot deniyor.
    print(f"Ad: {self.adi} soyadı: {self.soyadi} maaşı: {self.maasi}")

calisan1 = calisan('ali', 'demir', 5000, 32)
calisan2 = calisan('ahmet', 'yaprak', 3500, 62)

calisan.show_info(calisan1)
calisan1.show_info()
```

```python
class calisan:
  zam_orani = 1.1
  def __init__(self, adi, maasi):
    self.maasi = maasi

calisan1 = calisan('ali', 5000)
calisan2 = calisan('ahmet', 3500)

print(calisan1.zam_orani)  # bu şekilde erişebiliyoruz.
calisan.zam_orani = 1.2  # zam oranını güncelledim.
print(calisan2.zam_orani)
print(calisan.zam_orani)  # class'ın zam oranı. buda güncellendi.
```
