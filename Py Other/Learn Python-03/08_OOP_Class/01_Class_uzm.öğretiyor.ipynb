```python
class ayakkabi():
    def __init__(self, marka='', numara=0) -> None:
        self.markasi = marka
        self.numarasi = numara

    def ayakkabi_göster(self):
        return f'Ayakkabı markanız: {self.markasi}, Ayakkabı numaranız: {self.numarasi}'


ayakkabi1 = ayakkabi('adidas', 42)
ayakkabi2 = ayakkabi('nike', 37)
print(ayakkabi1.ayakkabi_göster())
print(ayakkabi2.ayakkabi_göster())
```

```python
class KareAlan():
    def __init__(self, yukseklik = 1, genislik = 1 ) -> None:
        self.yukseklik = yukseklik
        self.genislik = genislik
    def hesapla(self):
        return self.yukseklik * self.genislik

Kare1 = KareAlan(3,2)
print('karenin alanı: ', Kare1.hesapla())
```

```python
 class ogrenci():
    def __init__(self) -> None:
        self.isim = ""
        self.sınıf = ""
        self.numara = ""
        self.cinsiyet = ""
        self.dtarihi = 0
        self.devamsızlık = 0
    def ekle(self):
        self.isim = input('öğrenci ismini giriniz')
        self.sınıf = input('öğrenci sınıf giriniz')  
        self.numara = input('öğrenci numara giriniz')  
        self.cinsiyet = input('öğrenci cinsiyet giriniz')  
        self.dtarihi = int(input('öğrenci dtarihi giriniz'))
        self.devamsızlık = int(input('öğrenci devamsızlık giriniz'))
    def sil(self):
        del self
    def guncelle(self):
        self.sınıf = input('öğrenci sınıf giriniz')
        self.numara = input('öğrenci numara giriniz') 
        gelmediği_gun = int(input('öğrenci devamsızlık giriniz'))
        self.devamsızlık += gelmediği_gun
    def goster(self):
        return f"""öğrencinin, Adı: {self.isim} 
        Sınıfı: {self.sınıf} 
        Numarası: {self.numara} 
        Cinsiyeti: {self.cinsiyet} 
        Doğum Tarihi: {self.dtarihi} 
        Devamsızlık: {self.devamsızlık}"""

ogrenci1 = ogrenci()
ogrenci1.ekle()
print(ogrenci1.goster())
```

```python
class Galeri():
    """
    galeride kiralık araç bulunmakta, 
    len fonk ile km gösterilmekte.
    """
    def __init__(self, marka = "", model = "", renk = "") -> None:
        self.marka = marka
        self.model = model
        self.renk = renk
    def __str__(self) -> str:
        return f"""
        objemize özel temel özellikler,
        Marka:{self.marka}
        model = {self.model}
        renk= {self.renk}
        """
    def __len__(self):
        return 100

araba1_objesi = Galeri("kia", "Sportage", "yeşil")
print(len(araba1_objesi), 'km')
print(araba1_objesi.__doc__)
```

```python
class Kuslar():
    def __init__(self) -> None:
        pass
    def bilgi(self):
        return 'bazı kuşlar uçar bazıları uçamaz.'
class Kartallar(Kuslar):
    def __init__(self) -> None:
        pass
    def bilgi(self):
        return super(Kartallar, self).bilgi()

kartal1 = Kartallar()
print(kartal1.bilgi())
```
