```python
ailem = {'eşim': 'esra',      # Köşeli parantezi dict kümesine,
         'eşimin yaşı': '34', # Key ve Value değerleri eklemek içinde kullanabiliriz.
         'çocuklarım' : 3,}
ailem['kızım'] = 'Semra'
ailem['çocuklarım'] = 10      # value değerlerini de değiştirebiliyoruz.
print(ailem)
```

```python
ailem = {'eşim': 'esra',            # .update() yöntemi ile dict'e yeni ögeler ekleyebiliriz.
         'eşimin yaşı': '34',       # .update() return yapmaz arka planda ekleme yapar.
         'çocuklarım' : 3,}         # .update() in [] kullanarak iteem eklemeden bir farkı daha dikkat çekmiştir.
ailem.update({'kızım' : 'Semra'})   # .update() yönteminde hazır dict kurallarına uygun eleman tanımlayarak metoda eklenir.
ailem.update({'çocuklarım' : 30})   # .update() yöntemi ile value değerlerini de değiştirebiliyoruz.
print(ailem)
```

```python
ailem['kızım'] = 'kübra'  # Kızım value sini değiştirdik.
ailem
```

```python
ailem = {'eşim': 'esra',      # .items() ve diğer 2 metod yeni dizeler oluşturur.
         'eşimin yaşı': '34',
         'çocuklarım' : 3,}
print(ailem)
print(ailem.items())      #.items() Tüm ögelere, liste içerisinde tuple'ler oluşturdu.
print(ailem.keys())       #.keys() anahtarlara, key'lerden string listesi oluşturdu.
print(ailem.values())     #.values() değerlere erişiriz. value'lerden string listesi oluşturdu.
```

```python
ailem = {'eşim': 'esra',      # Ayrıca del işlevini kullanarak bir öğeyi kaldırabilirsiniz
         'eşimin yaşı': '34', # Formül sözdizimi şöyledir: del sözlük_adı['anahtar']
         'çocuklarım' : 3,
         'kızım' : 'Semra'}
del ailem['kızım'], ailem['çocuklarım']  # birden fazla item'ı sildik.
ailem
```

```python
ailem = {'eşim': 'esra',      # in ve not in operatörünü kullanarak, 
         'eşimin yaşı': '34', # anahtarın sözlükte olup olmadığını kontrol edebilirsiniz.
         'çocuklarım' : 3,}   
print('çocuklarım' in ailem.keys())
print('kızım' not in  ailem)  # kızım'ı silmiştik.
print('esra' in ailem)        # Value yi sorgulamak istiyorsak values() metodunu kullanmalıyız.
print('esra' in ailem.values())
```
