```python
numbers = [-1, 3, 4, 7, 3, 0, 3, 16, 3, 7, 0, 0, 1]
```

```python
max(numbers)  # min, max 'ın default' u parantez içerisine yazılanın en büyüğü veya en küçügünü bulmak.
max(numbers, key=numbers.count) # default'una key atadık ve dedik ki; numberste tüm elemanlara bak en çok tekrar edeni döndür.
```

```python
numbers.count(max(numbers, key=numbers.count))  # burada da numberste tüm elemanlara bak en çok tekrar edeni bul. kaç tekrar ettiyse onu döndür.
```
