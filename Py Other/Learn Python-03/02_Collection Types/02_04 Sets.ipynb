
* .add(): Sete yeni bir öğe ekler.
* .remove(): Bir öğeyi silmemizi sağlar.
* .intersection(): İki kümenin kesişimini verir. (&)
* .union(): İki kümenin birleşimini döndürür (|)
* .difference(): İki kümenin farkını alır. (-)

```python
set('nuriye vvv')
```

```python
a = set('nuriye')
b = set('nuri')
print(a-b)   # difference metodu ile aynı sonucu verir.
print(a.difference(b))
```

```python
a = set('nuriye')
b = set('nuri')
print(a|b)   # union metodu ile aynı sonucu verir.
print(a.union(b))
```

```python
a = set('nuriye')
b = set('nuri')
print(a&b)   # intersection metodu ile aynı sonucu verir.
print(a.intersection(b))
```

```python
a = set('nuriye')
a.remove('i')  # 'i' yi sildik.
a
```

```python
a = set('nuriye')
a.add('isim')  
a
```
