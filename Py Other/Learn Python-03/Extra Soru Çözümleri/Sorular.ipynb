````python
# Kurallar çerçevesinde parola oluşturma kodu yazalım.
def checkPassword(psw):
    import re
    if len(psw)<8:
        raise Exception('Parola en az 7 karakterli olmalıdır.') # kullanıcı belirlenen şartlara uymazsa, bir hata mesajı gösterilip programdan çıkılacaktır.
    elif not re.search("[a-z]",psw):
        raise Exception('Parola küçük harf içermelidir.')
    elif not re.search("[A-Z]",psw):
        raise Exception('Parola büyük harf içermelidir.')
    elif not re.search("[0-9]",psw):
        raise Exception('Parola rakam içermelidir.')
    elif re.search(" " ,psw): # re.search metodu ile boşluk sorgulaması yaptık. 
        raise Exception('Parola boşluk içermemelidir.')
    # else:
    #     print('geçerli parola')
password=input('paralo gir')
try:  # bu blokta da checkPassword fonksiyonunda hata varsa except bloğunu çalıştır. hata yoksa else blogu çalışır.
    checkPassword(password)
    
except Exception as ex:
    print(ex)
else:
    print('parola geçerli')
````

# Verilen sayı boyutunda bir alfabetik rangoli yazdıralım. (Rangoli, desenlerin yaratılmasına dayanan bir Hint halk sanatı biçimidir.)

````python
def print_rangoli(size):
  width = size * 4 - 3  # genişliğini ayarlıyoruz.
  string = ''
  for i in range(1, size+1):  # size kadar döngü oluşturuyoruz.
    for j in range(0, i):
      string += chr(96+size-j) 
      if len(string) < width:
        string += '-'
    for k in range(i-1, 0, -1):
      string += chr(97+size-k)
      if len(string) < width :
        string += '-'
    print(string.center(width, '-'))
    string = '' 

  for i in range(size-1, 0, -1):  # döngüyü geriye doğru çalıştırıyoruz.
    string = ''
    for j in range(0, i):
      string += chr(96+size-j)
      if len(string) < width:
        string += '-'
    for k in range(i-1, 0, -1):
      string += chr(97+size-k)
      if len(string) < width :
        string += '-'
    print(string.center(width, '-'))

print_rangoli(int(input('Bir sayı giriniz: ')))
````

leap Year 

Genel bir kural olarak, artık yıllar 4 rakamının katı olan yıllardır:

1980, 1984, 1988, 1992, 1996, 2000, 2004, 2008, 2012, 2016, 2020, 2024, 2028, 2032, 2036, 2040, 2044, 2048 vb.
Ancak bu kuralın iki istisnası vardır:

100'ün katı olan yıllardan sadece 400'e kalansız olarak bölünebilenler artık yıldır:

Örneğin 1600 ve 2000 yılları artık yıldır ancak 1800 ve 1900 artık yıl değildir.

````python
def CheckLeap(Year):
  # Checking if the given year is leap year  
  if((Year % 400 == 0) or
     (Year % 100 != 0) and
     (Year % 4 == 0)):
    print("Given Year is a leap Year");
  # Else it is not a leap year  
  else:
    print ("Given Year is not a leap Year")
# Taking an input year from user  
Year = int(input("Enter the number: "))
# Printing result  
CheckLeap(Year)
````

asigment 3 bir Armstrong numarasımı githup a girilecek.

````python
n = input('bir sayı giriniz: ')
u = len(n)
s = 0
total = 0

if n > '0' and n.isdigit() == True :
  while s < u :
    for i in n :
        i = int(n[s])**u
        s +=1
        total = total + i
    if total == int(n) :
      print(total, 'bir Amstrong numarasıdır')
    else :
      print(n, 'bir Amstrong numarası değildir.')
else:
  print('Geçersiz bir giriş. Sayısal olmayan, flaot veya negatif değerler kullanmayın!')
````

````python
while True :
    number = input("Enter a positive integer number :")
    digit = len(number)
    summ = 0

    if int(number) >= 0 :
      for i in range(digit):
        summ = summ + int(number[i]) ** digit

      if summ == int(number):
        print(number, " is an Amstrong Number")
        break
      else:
        print(number, " is NOT an Amstrong Number")
        break
````

Kullanıcıdan binary bir numara girmesini isteyelim ve onu decimal sayıya convert edelim örn: kullanıcı 1101001 sayısını girdi çıktı : 105 olmalı

````python
binary_number = input('lütfen binary bir numara giriniz: ')
x = len(binary_number)
y = 0

while x :
  for i in binary_number:
   y += int(i) * 2**(x-1)
   x-=1
  print(y)
````

Aşağıdaki 2 soru aynı sonucu elde eder ve bir listedeki frekans değeri ve bu değerin distedeki miktarını verir. 

````python
from collections import Counter
numbers = [1, 3, 7, 4, 3, 0, 3, 6, 3]
a = Counter(numbers)  # Counter() metodu ile önce her elemanın frekansını bulduk.
b = a.most_common(1)  # bulduğumuz frekans değerlerine most_common(1) metodu uygulayarak en yüksek frekanslı elemanı (2 elemanlı Tuple) bulduk.
text = "the most frequent number is {} and it was {} times repeated".format(b[0][0], b[0][1]) # format metodu ve içerisinde indexleme yaparak sonucu istenilen formata dönüştürdük.
print(text)
````

````python
numbers = [1, 3, 7, 4, 3, 0, 3, 6, 3]
a = max(numbers, key = numbers.count)  # max fonksiyonunun default'u uygulandığı listedeki en büyük değeri döndürmesidir. ancak key atayarak max frekans bulma işlemi yaptırdık.
b = numbers.count(max(numbers, key = numbers.count))  # yani key ile listedeki tüm sayıları say en çok olanı döndür demiş olduk.
text = "the most frequent number is {} and it was {} times repeated".format(a, b) 
print(text)
````
