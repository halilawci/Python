# 7 -   Kullanıcının klavyeden girdiği sayı kadar öğrencinin vize1,vize2 ve final notlarını kullanıcı girişi ile yapınız.   

- Daha sonra girilen öğrencilerin aritmetik not ortalamasını alınız
- Ekrana 70 ve üzerinde ortalamaya sahip öğrencileri yazdırınız.


Hedefler:

- Döngüler ile çalışabilme
- Dictionaryleri kullanabilme, key ve value değerlerine ulaşabilme
- Listeleri kullanabilme ve üzerinde işlem yapabilme
- Aritmetik operatörleri kullanabilme

```python
number=int(input('kaç adet öğrencinin notlarını gireceksiniz: '))
i=1
dict_table={}


while i<(number+1): 
  not_list=[]
  name = input(str(i)+'.öğrencinin ismini giriniz: ')
  
  not_list.append(int(input(str(i)+'.öğrencinin 1.vize notunu giriniz: ')))
  not_list.append(int(input(str(i)+'.öğrencinin 2.vize notunu giriniz: ')))
  not_list.append(int(input(str(i)+'.öğrencinin final notunu giriniz: ')))
  dict_table.update({name:not_list})
  i+=1

print(dict_table)

for key in dict_table:
   #print(sum(dict_table[key])/3)
   if sum(dict_table[key])/3 >= 70:
      print(key, "{:.2f}".format(sum(dict_table[key])/3))
```

```python
students = dict()
num_student = int(input("For how many students will you enter grades? "))
student_name_pass = []
student_name_fail = []
for i in range(num_student):
    student_name = input("Enter names of student :")
    grades= []
    midterm_1 = float(input(f"Enter 1. Midterm of {student_name} :"))
    grades.append(midterm_1)
    midterm_2 = float(input(f"Enter 2. Midterm of {student_name} :"))
    grades.append(midterm_2)
    final = float(input(f"Enter Final exam of {student_name} :"))
    grades.append(final)
    students[student_name] = grades
    grade_average = midterm_1*(0.3) + midterm_2*(0.3) + final*(0.4)
    if grade_average > 70 :
        student_name_pass.append(student_name)
        print(f" {student_name} pass with {grade_average} \n")
    else:
        student_name_fail.append(student_name)
        print(f" {student_name} fail with {grade_average}\n")
print("Dictionary of student created : \n")
print(students, "\n\n")
print("Pass List", student_name_pass)
print("Fail List", student_name_fail )
```

# 8 -   Girilen bir texti (string) şifrelemek için aşağıdaki adımları gerçekleştireceğim ve çıktı olarak 'şifre : 010010011100' şeklinde bir çıktı alacağım.

- Girilen stringin tek tek karakterleri ASCII koda çevrilecek
- ASCII kod karşılıkları 2lik (ikilik - binary) tabana çevrilecek
- son olarak 2 lik taban karşılıkları yanyana yazdırılarak bir sayı veya metin üretilecek

Hedefler:

- Döngüler ile çalışabilme
- içiçe döngüleri kullanabilme (nested loops in python)
- string değişkenler üzerinde işlem yapabilme
- Bir karakterin ASCII koduna ulaşabilme ( ord() metodunu kullanabilme )
- 10 tabanındaki bir sayıyı 2 lik tabana çevirebilme
- Aritmetik operatörleri kullanabilme 
- (opsiyonel olarak) dizileri kullanabilme ve üzerinde işlemler yapabilme


sifre.jpg

```python
text=input('Kullanıcı adınızı giriniz: ')
password=[]
for i in text:
  password.append(ord(i))

print(password)

pass_text=''
for j in password:
  binary=''
  while j > 0 :
    kalan = j % 2  # 2'ye bölümden kalan.. 
    binary=str(kalan)+binary  # değer girildikçe str değişkenimin ilk elemanı olacak.
    j //= 2 

  print(binary) #herbir ascii değerin 2lik tabandaki karşılığını ekrana yazdırıyorum. adım adım görelim diye
  
  pass_text+=binary #tüm binaryleri tek bir string te birleştiriyorum. 

print('girilen metnin şifreli hali : ',pass_text)
```

```python
x=int(input('bir sayı giriniz : '))
top=0

for i in range(1,x):
  if x%i==0:
    top+=i

if top==x:
  print(x,' sayısı mükemmel bir sayıdır.')
else:
  print(x,' sayısı mükemmel bir sayı değildir.')
```

```python
#selection sort algoritması kullanılarak çözülmüştür.

listem=[3, 5, 1, 0, 6, 10, 2, 9, 4, 15, 25]

print(listem)  # ekrana yazdırıp iki liste arasında ki fark görülsün.

for i in range(0,len(listem)):
    enk=listem[i]
    index=i
    for j in range(i,len(listem)):
        if listem[j]<enk:
            enk=listem[j]
            index=j
            
    listem[index]=listem[i]
    listem[i]=enk


print(listem)
```
