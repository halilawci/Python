Size boş olmayan integerlardan oluşan bir liste veriliyor. Bu görev için içinde unique eleman olmayan bir liste return etmeniz isteniyor. Bunu yapmak için size verilen listede sadece bir kere geçen elemanları kaldırmanız gerekmektedir. Bunu yaparken listenin sırasını değiştimemeniz gerekmektedir. Lütfen örnekleri inceleyiniz. 

```python
list = [1, 2, 3, 4, 5, 5, 4]
boş = []
for i in list:
  if list.count(i)>1:
    boş.append(i)
print(boş)
```

Sizden kelimenin bir parçası olmayan sayıları toplamanız istenmektedir.  Lütfen örnekleri inceleyiniz. 

```python
string = 'Petersen between 1845 and 1910 year'
toplam = 0
for i in string.split():
  if i.isdigit():
    toplam += int(i)

print(toplam)
```
