You work for a manufacturer as a programmer and have been asked to calculate the total profit made on the sales of a product.

```python
products = {
'cost': int(input('enter cost: ')),
'sales_value': int(input('Enter the sale price: ')),
'freq':int(input('Enter the inventory amount: '))
}
total_profit = round((products['sales_value']-products['cost'])*products['freq'])
print(total_profit)
```
